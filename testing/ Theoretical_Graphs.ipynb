```python
import matplotlib.pyplot as plt
```

```python
def get_pi(i, l, m, N):
    if l != m:
        return (l/m)**i * (1 - l/m)/(1-(l/m)**(N+1)) 
    else:
        return 1/(N+1)
```

```python
def get_expect(l, m, N):
    exp = 0
    for i in range(N+1):
            exp += i * get_pi(i, l, m, N)
    return exp
```

```python
def get_expect2(l, m, N):
    r = l/m
    if r != 1:
        return (r/(1-r) - (N+1)*(r**(N+1))/(1-r**(N+1)))
    else:
        print("something")
```

```python
def get_average_time(l, m, N):
    return (1/(l*(1-get_pi(N, l, m, N))) * get_expect(l, m, N))
```

```python
m = [10, 100, 3, 1000]
N = [20, 5, 1000, 10]
```

```python
plt.figure(figsize=(15, 10))
plt.suptitle('1.       lambda')

for i in range(len(m)):
    lamb = range(m[i]-(m[i]//2+1), m[i]+m[i]//2+2)
    pN = []
    for l in lamb:
        pN.append(get_pi(N[i], l, m[i], N[i]))

    plt.subplot(221+i)    
    plt.plot(lamb, pN, 'o-')
    plt.title('mu = {}, N = {}'.format(m[i], N[i]))
    plt.xlabel('lambda')
    plt.ylabel('p_N')
    plt.grid()

plt.show()    
```

```python
plt.figure(figsize=(15, 10))
plt.suptitle('2.         lambda')

for i in range(len(m)):
    lamb = range(m[i]-(m[i]//2+1), m[i]+m[i]//2+2)
    expect = []
    for l in lamb:
        expect.append(get_expect(l, m[i], N[i]))

    plt.subplot(221+i)    
    plt.plot(lamb, expect, 'o-')
    plt.title('mu = {}, N = {}'.format(m[i], N[i]))
    plt.xlabel('lambda')
    plt.ylabel('average number of items')
    plt.grid()

plt.show()  
```

```python
plt.figure(figsize=(15, 10))
plt.suptitle('3.          lambda')

for i in range(len(m)):
    lamb = range(m[i]-(m[i]//2+1), m[i]+m[i]//2+2)
    av_time = []
    for l in lamb:
        av_time.append(get_average_time(l, m[i], N[i]))

    plt.subplot(221+i)    
    plt.plot(lamb, av_time, 'o-')
    plt.title('mu = {}, N = {}'.format(m[i], N[i]))
    plt.xlabel('lambda')
    plt.ylabel('average time')
    plt.grid()

plt.show()  
```

```python
plt.figure(figsize=(15, 10))
plt.suptitle('4.       lambda')

for i in range(len(m)):
    lamb = range(m[i]-(m[i]//2+1), m[i]+m[i]//2+2)
    p_wait = []
    for l in lamb:
        p_wait.append(1 - get_pi(0, l, m[i], N[i]))

    plt.subplot(221+i)    
    plt.plot(lamb, p_wait, 'o-')
    plt.title('mu = {}, N = {}'.format(m[i], N[i]))
    plt.xlabel('lambda')
    plt.ylabel('probability of waiting')
    plt.grid()

plt.show()  
```
